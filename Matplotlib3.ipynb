```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\miner\Python_Matplotlib\googleplaystore.csv',nrows=1000)# Reads the first 1000 rows of the specified CSV file into a pandas DataFrame
df
```

```python
x = df['Rating']  # Extracts the 'Rating' column from the DataFrame and assigns it to variable x
y = df['Reviews']  # Extracts the 'Reviews' column from the DataFrame and assigns it to variable y
```

```python
plt.scatter(x, y)  # Creates a scatter plot using the 'Rating' values (x) and 'Reviews' values (y)
plt.show()  # Displays the scatter plot on the screen
```

```python
plt.scatter(x, y)  # Creates a scatter plot with 'Rating' on the x-axis and 'Reviews' on the y-axis
plt.title("Google Play Store Apps Scatter plot")  # Adds a title to the plot
plt.xlabel("Rating")  # Labels the x-axis as 'Rating'
plt.ylabel("Reviews")  # Labels the y-axis as 'Reviews'
plt.show()  # Displays the plot on the screen
```

```python
plt.scatter(x, y, c='r')  # Creates a scatter plot with red ('r') color for the points
plt.title("Google Play Store Apps Scatter plot")  # Adds a title to the plot
plt.xlabel("Rating")  # Labels the x-axis as 'Rating'
plt.ylabel("Reviews")  # Labels the y-axis as 'Reviews'
plt.show()  # Displays the plot on the screen
```

```python
plt.figure(figsize=(16, 9))  # Sets the size of the figure to 16x9 inches
plt.scatter(x, y, c='r')  # Creates a scatter plot with red ('r') color for the points
plt.title("Google Play Store Apps Scatter plot")  # Adds a title to the plot
plt.xlabel("Rating")  # Labels the x-axis as 'Rating'
plt.ylabel("Reviews")  # Labels the y-axis as 'Reviews'
plt.show()  # Displays the plot on the screen
```

```python
plt.figure(figsize=(16, 9))  # Sets the size of the figure to 16x9 inches
plt.scatter(x, y, c='r', marker="*", s=100, alpha=0.5)  # Creates a scatter plot with red ('r') color, star markers ('*'), size 100, and 50% transparency (alpha=0.5)
plt.title("Google Play Store Apps Scatter plot")  # Adds a title to the plot
plt.xlabel("Rating")  # Labels the x-axis as 'Rating'
plt.ylabel("Reviews")  # Labels the y-axis as 'Reviews'
plt.show()  # Displays the plot on the screen
```

```python
plt.figure(figsize=(16, 9))  # Sets the size of the figure to 16x9 inches
plt.scatter(x, y, c='r', marker="*", s=100, alpha=0.5, linewidths=10)  # Creates a scatter plot with red ('r') color, star markers ('*'), size 100, 50% transparency (alpha=0.5), and line width of 10 around the markers
plt.title("Google Play Store Apps Scatter plot")  # Adds a title to the plot
plt.xlabel("Rating")  # Labels the x-axis as 'Rating'
plt.ylabel("Reviews")  # Labels the y-axis as 'Reviews'
plt.show()  # Displays the plot on the screen
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c='r',marker="*",s=100,alpha=0.5,linewidths=10,edgecolors="g")
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```
